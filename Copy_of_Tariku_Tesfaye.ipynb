```python
# ======= Breast Cancer: full pipeline (Colab-ready) =======
# Choose dataset source: "sklearn" for built-in dataset, or "csv" if you uploaded a CSV to /content/dataset/
DATA_SOURCE = "sklearn"  # options: "sklearn" or "csv"
CSV_PATH = "/content/dataset/breast_cancer.csv"  # change if using CSV

# ----------------- Imports -----------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc,
                             log_loss, classification_report)
import warnings
warnings.filterwarnings("ignore")
sns.set(style="whitegrid")

# ----------------- Load data -----------------
if DATA_SOURCE == "sklearn":
    from sklearn.datasets import load_breast_cancer
    bc = load_breast_cancer()
    X = pd.DataFrame(bc.data, columns=bc.feature_names)
    y = pd.Series(bc.target, name="target")  # 0 = malignant, 1 = benign
    df = pd.concat([X, y], axis=1)
    print("Loaded sklearn breast_cancer dataset. Shape:", df.shape)
else:
    # CSV: ensure you uploaded and unzipped into /content/dataset/
    df = pd.read_csv(CSV_PATH)
    print("Loaded CSV from:", CSV_PATH, " Shape:", df.shape)
    # If target column name is unknown, try to infer common names:
    if "target" in df.columns:
        y = df["target"]
        X = df.drop("target", axis=1)
    elif "diagnosis" in df.columns:
        # common Kaggle column: 'diagnosis' with values M/B or 0/1
        # Map if necessary
        if df["diagnosis"].dtype == object:
            df["diagnosis"] = df["diagnosis"].map({"M":0, "B":1})
        y = df["diagnosis"]
        X = df.drop("diagnosis", axis=1)
    else:
        # fallback: assume last column is target
        y = df.iloc[:, -1]
        X = df.iloc[:, :-1]

# ----------------- Quick EDA -----------------
print("\n--- HEAD ---")
display(df.head())
print("\n--- INFO ---")
print(df.info())
print("\n--- DESCRIPTION (numerical cols) ---")
display(df.describe().T)

print("\n--- Missing values per column ---")
print(df.isnull().sum())

print("\n--- Target distribution ---")
display(y.value_counts())
sns.countplot(x=y)
plt.title("Target class distribution (0=malignant, 1=benign)")
plt.show()

# ----------------- Correlation heatmap (top features) -----------------
# Show correlation between features and the target (pearson)
corr_with_target = df.corr().abs()["target"].sort_values(ascending=False) if "target" in df.columns else pd.Series()
if not corr_with_target.empty:
    top_feats = corr_with_target.index[1:16]  # top 15 features correlated with target (skip 'target' itself)
    plt.figure(figsize=(10,8))
    sns.heatmap(df[top_feats].corr(), annot=False, cmap="coolwarm")
    plt.title("Correlation among top features")
    plt.show()

# ----------------- Preprocessing -----------------
# If there are categorical columns, encode them here (this dataset is numeric)
# Fill missing values if any:
if X.isnull().sum().sum() > 0:
    print("Filling missing numeric values with column median.")
    X = X.fillna(X.median())

# Standardize features (recommended for many models)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# ----------------- Train-test split -----------------
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y)

print("\nTrain shape:", X_train.shape, "Test shape:", X_test.shape)

# ----------------- Utility: evaluation function -----------------
def evaluate_model(model, X_test, y_test, name="model"):
    y_pred = model.predict(X_test)
    y_proba = None
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    ll = log_loss(y_test, y_proba) if y_proba is not None else None

    print(f"\n=== Evaluation: {name} ===")
    print("Accuracy:  {:.4f}".format(acc))
    print("Precision: {:.4f}".format(prec))
    print("Recall:    {:.4f}".format(rec))
    print("F1-score:  {:.4f}".format(f1))
    if ll is not None:
        print("Log Loss:  {:.4f}".format(ll))

    print("\nClassification report:\n", classification_report(y_test, y_pred, zero_division=0))
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["malignant(0)","benign(1)"])
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix: {name}")
    plt.show()

    # ROC curve & AUC (if probabilistic)
    if y_proba is not None:
        # For binary: take probability of class 1
        if y_proba.shape[1] == 2:
            probs = y_proba[:, 1]
        else:
            probs = y_proba[:, 0]
        fpr, tpr, _ = roc_curve(y_test, probs)
        roc_auc = auc(fpr, tpr)
        plt.figure()
        plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.3f})')
        plt.plot([0,1],[0,1],'--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve: {name}')
        plt.legend(loc='lower right')
        plt.show()

# ----------------- 1) Decision Tree (Gini) -----------------
dt_gini = DecisionTreeClassifier(criterion="gini", max_depth=4, random_state=42)
dt_gini.fit(X_train, y_train)
evaluate_model(dt_gini, X_test, y_test, name="Decision Tree (Gini)")

# Visualize tree
plt.figure(figsize=(20,10))
plot_tree(dt_gini, feature_names=X.columns, class_names=["malignant","benign"], filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree (Gini) - max_depth=4")
plt.show()

# Feature importance
fi = pd.Series(dt_gini.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(8,6))
fi.head(15).plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Top 15 Feature Importances (Gini Tree)")
plt.xlabel("Importance")
plt.show()

# ----------------- 2) Decision Tree (Entropy) -----------------
dt_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=42)
dt_entropy.fit(X_train, y_train)
evaluate_model(dt_entropy, X_test, y_test, name="Decision Tree (Entropy)")

plt.figure(figsize=(20,10))
plot_tree(dt_entropy, feature_names=X.columns, class_names=["malignant","benign"], filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree (Entropy) - max_depth=4")
plt.show()

fi2 = pd.Series(dt_entropy.feature_importances_, index=X.columns).sort_values(ascending=False)
plt.figure(figsize=(8,6))
fi2.head(15).plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Top 15 Feature Importances (Entropy Tree)")
plt.xlabel("Importance")
plt.show()

# ----------------- Cross-validated scores (for both criteria) -----------------
from sklearn.model_selection import cross_validate
cv_results_gini = cross_validate(DecisionTreeClassifier(criterion="gini", max_depth=4, random_state=42),
                                 X_scaled, y, cv=5, scoring=["accuracy","precision","recall","f1"])
cv_results_entropy = cross_validate(DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=42),
                                    X_scaled, y, cv=5, scoring=["accuracy","precision","recall","f1"])

print("\nCross-validated metrics (mean ± std) - Gini")
for k in ["test_accuracy","test_precision","test_recall","test_f1"]:
    v = cv_results_gini[k]
    print(f"{k}: {v.mean():.4f} ± {v.std():.4f}")

print("\nCross-validated metrics (mean ± std) - Entropy")
for k in ["test_accuracy","test_precision","test_recall","test_f1"]:
    v = cv_results_entropy[k]
    print(f"{k}: {v.mean():.4f} ± {v.std():.4f}")

# ----------------- Optional: tune max_depth quickly to see effect -----------------
depths = range(1,11)
scores = []
for d in depths:
    m = DecisionTreeClassifier(criterion="entropy", max_depth=d, random_state=42)
    cv = cross_val_score(m, X_scaled, y, cv=5, scoring="accuracy")
    scores.append(cv.mean())

plt.figure(figsize=(8,5))
plt.plot(depths, scores, marker="o")
plt.xlabel("max_depth")
plt.ylabel("CV accuracy")
plt.title("Depth vs CV Accuracy (entropy)")
plt.xticks(depths)
plt.show()

# ========== End of pipeline ==========
print("Pipeline finished. You can change DATA_SOURCE and CSV_PATH to use a different file.")
```
